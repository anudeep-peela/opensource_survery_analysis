# Data

To explore programmers characteristics. I used the Stack Overflow Developer Survey Results 2019

```python
sr_2019 = pd.read_csv('2019_survey_results_public.csv')
sr_2019.shape
```

## pre processing steps

```python
# Defining Opensourcer as the programmer how contributes more than once a year
```

```python
sr_2019['OpenSourcer'].replace('Once a month or more often', 'Yes', inplace = True)
sr_2019['OpenSourcer'].replace('Less than once a month but more than once per year', 'Yes', inplace = True)
sr_2019['OpenSourcer'].replace('Less than once per year', 'No', inplace = True)
sr_2019['OpenSourcer'].replace('Never', 'No', inplace = True)
```

```python
(100*sr_2019['OpenSourcer'].value_counts()/88883).plot(kind = 'bar')
plt.title('OpenSource vs Non-OpenSource Programmers Distribution')
plt.savefig('distributions')
```

```python
(100*sr_2019['OpenSourcer'].value_counts()/88883)
```

Hmm, there are about 31.98 Per cent Open source contributors.

2. Let's look at salary distributions of programmers

```python
#compenstation earned by opensourcer and non-opensourcer
```

```python
sr_2019.groupby('OpenSourcer')['ConvertedComp'].mean().plot(kind='bar')
plt.title('OpenSource vs Non-OpenSource Programmers Salary Distribution')
plt.savefig('salary_distributions')
```

```python
sr_2019.groupby('OpenSourcer')['ConvertedComp'].std()
```

Awesome. It seems that opensource programmer is paid on average $7465 more than Non-open source programmer.

Let us dig the data to find out the practises used by programmers. From the data, I found out Code reviews and Unit testing features in the data set.

## Code Reviews
Code review is a software quality assurance activity in which one or several people check a program mainly by viewing and reading parts of its source code.
Code reviews usually help in better code quality, knowledge transfer.

```python
## BarPlot of Opensourcer vs CodeRevHrs spent on an average
sr_2019.groupby('OpenSourcer')['CodeRevHrs'].mean().plot.barh()
plt.title('Code Reviews Distribition')
plt.savefig('coderevs')
```

From the data, It seems like Opensource contributor spends 50 mins more in code reviewing than a Non-Opensource contributor.
Lets us see their opinions and compensation

```python
## BarPlot of Opensourcer vs Code Reviews opinions spent on an average
sr_2019.groupby('CodeRev')['ConvertedComp'].mean().plot(kind='bar')
plt.title('Code Reviews Opinions vs Compensation Distribition')
plt.savefig('coderevs_op',bbox_inches = 'tight')
```

The programmer who sees the value in code reviews are receiving better pay.

## Unit Testing:
unit testing is a software testing method by which individual units of source code is tested to determine whether they are fit for use.
Now, let's explore unit testing practices.

```python
## BarPlor of Compenstaed salary vs Unit testing opinions
sr_2019.groupby('UnitTests')['ConvertedComp'].mean().plot(kind='bar')
plt.title('Unit Testing Opinions vs Compensation Distribition')
plt.savefig('unit_op',bbox_inches = 'tight')
```

It seems like programmers who do unit testing and consider it as a part of process receive better compensations.

# Conclusion

From the above analysis, it is evident that open source contributors developer are paid better and the programmer who sees the value in code reviews and unit testing are better paid.
